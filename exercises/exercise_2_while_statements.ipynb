### Example - Oil leakage

There is an oil leakage in the bird population in an island to 1/2 in each year. From start there were 80,000 birds. How many years does it take for it to have 1/10 remaining?

```python
birds = 80_000
year = 0

while birds > 8000:
    print(f'Year {year} there were {birds} birds')
    birds /= 2
    year += 1

print (f'It takes {year} years for the birds tio have 1/10th of population')
```

## Break statement

```python
import random as rnd
while True:
    number1 = rnd.randint(1, 10)
    number2 = rnd.randint(1, 10)

    user_answer = int(input(f'What is {number1}*{number2}'))
                            
    if user_answer == number1 * number2:
        print("correct")
    else:
        print(f'Wrong, the answer was {number1 * number2}')


    play_again = input("Wanna play again? (y / n)")
    
    if play_again != "y":
        print("Have a nice day")
        break
```

## While loops and the break statement (from other videos)

```python
# start with 0
total = 0  

# loop through the numbers 1 up to (but not including 5) -> 1, 2, 3, 4
for i in range (1, 5):

    # add the current number to total
    total += i
    print(total)
```

```python
 # start with 0
total2 = 0

#start counting form 1
j = 1

# repeat while j is less than 5
while j < 5:

    # add the current value of j to total2
    total2 += j

    # increases j by 1 so the loop moves forward
    j += 1
    print (total2)
```
