## while statements

### Example - Oil leakage

There is an oil leakage in the bird population in an island to 1/2 in each year. From start there were 80,000 birds. How many years does it take for it to have 1/10 remaining?

```python
birds = 80_000
year = 0

while birds > 8000:
    print(f'Year {year} there were {birds} birds')
    birds /= 2
    year += 1

print (f'It takes {year} years for the birds tio have 1/10th of population')
```

## Break statement

```python
import random as rnd
while True:
    number1 = rnd.randint(1, 10)
    number2 = rnd.randint(1, 10)

    user_answer = int(input(f'What is {number1}*{number2}'))
                            
    if user_answer == number1 * number2:
        print("correct")
    else:
        print(f'Wrong, the answer was {number1 * number2}')


    play_again = input("Wanna play again? (y / n)")
    
    if play_again != "y":
        print("Have a nice day")
        break
```
