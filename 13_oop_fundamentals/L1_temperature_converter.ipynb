### üß† Hint 2 ‚Äì Conversion formulas

You can always convert from any unit to Celsius first ‚Äî then build the others off that.
Use the known relationships:

From Celsius to Fahrenheit: F = (C * 9/5) + 32

From Fahrenheit to Celsius: C = (F - 32) * 5/9

From Celsius to Kelvin: K = C + 273.15

From Kelvin to Celsius: C = K - 273.15

That means you can write one method, say to_celsius(), that checks what your current unit is and converts it.
