### Summary when to use tuples vs list

<div style="max-width:66ch;">

        <thead>
            <tr>
                <th>Feature</th>

        <tbody>
            <tr>
                <td><strong>Mutability</strong></td>

            <tr>
                <td><strong>Use Cases</strong></td>

            </tr>
            <tr>
                <td><strong>Performance / Memory</strong></td>
                <td>Generally fine; slightly more overhead for mutability.</td>
                <td>Slightly faster & more memory efficient due to fixed size.</td>
            </tr>
            <tr>
                <td><strong>Dictionary Keys</strong></td>
                <td>Cannot be used as dictionary keys (not hashable).</td>
                <td>Can be used as dictionary keys (hashable if contents are).</td>
            </tr>
            <tr>
                <td><strong>Semantic Intent</strong></td>
                <td>Implies a collection that may change.</td>
                <td>Implies a fixed, ordered set of related items.</td>
            </tr>

2. Counting and Searching

Make a tuple of repeated values (e.g., colors).

Count how many times "red" appears.

Find the index of the first "blue".

```python
colors = ("red", "blue", "green", "red", "blue", "green", "red", "blue", "green", "red", "blue", "green")
red_count = colors.count("red")
blue_count = colors.count("blue")
print(f'Red count: {red_count}\nBlue_count: {blue_count}')
first_blue = colors.index("blue")
print(f'First blue index: {first_blue}')
    
```

```python
# for loop version
colors = ("red", "blue", "green", "red", "blue", "green", "red", "blue", "green", "red", "blue", "green", "blue")

red_count = 0
blue_count = 0

for x in colors:
    if x == "red":
        red_count += 1
    if x == "blue":
        blue_count +=1

print(red_count)
print(blue_count)


```

3. Nested Tuples

Create a tuple of 3 students, where each student is represented as a tuple (name, age).

Print only the names.

Print the average age.

```python
from statistics import mean

# create tuple for each student
a = ("aira", 29)
b = ("aya", 37)
c = ("mark", 41)

# place in one big tuple
students = (a, b, c)   #packing

for student in students:
    name, age = student      #unpacking
    print(f'name: {name}\n age: {age}\n...........')


# get the average age
ave_age = int(sum(age for _, age in students) / len(students))
print(f'Average age: {ave_age}')
```

4. Tuple from User Input

Ask the user to enter 5 numbers separated by spaces.

Store them as a tuple of integers.

Print the tuple.

```python
# Ask user for numbers
numbers = input("Enter numbers separated by commas: ")

# Step 1: convert to a tuple of integers
original_tuple = tuple(int(x) for x in numbers.split(","))

print("Original tuple:", original_tuple)

# Step 2: convert tuple to list and sort
sorted_list = sorted(original_tuple)

# Step 3: convert sorted list back to tuple
sorted_tuple = tuple(sorted_list)

print("Sorted tuple:", sorted_tuple)
```

5. Tuple Sorting (with Conversion)

Create a tuple of numbers.

Convert it into a list, sort it, and convert it back into a tuple.

Print before and after.

```python
numbers = (31, 52, 65, 86, 42)

list_numbers = list(int(x) for x in numbers)  # one line for loop convertion to int and list
print (f' List: {list_numbers}')
tuple_numbers = tuple(sorted(list_numbers))
print(f'Tuple: {tuple_numbers}')
```
